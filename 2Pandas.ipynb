```python
coffee.isna().sum() # indetail details about the null values in ech column
```

```python
# coffee.fillna(coffee['Coffee Type'].mean()) # Cannot calculate mean of string values

# Fill null values with the mode (most frequent value)
coffee['Coffee Type'] = coffee['Coffee Type'].fillna(coffee['Coffee Type'].mode()[0])
coffee
```
