## Logistic Regression
Logistic regreesion is for the Classification tasks, it fits a curve tothe data that has either 0 or 1 as the outputs, the function for the curve used to fit is the sigmoid function, for the value of x(input) the sigmoid function returns a value betweeen 0 or 1, with the threshold of 0.5, the probability of that prediction is 1 if greater than o.5, and probability of that prediction is 0 if that is less than 0.5

sigmoid function

g(z) = $1/1+e^{-z}$

for a large  +ve value of z, the sigmoid(logistic) function returns value closer to 1 (1/(closer to 1) ~ 1).

for a large  -ve value of z, the sigmoid(logistic) function returns value closer to 0 (1/(larger than 1) ~ 0).

## Decision Boundary

It's the Boundary that defines different categories

**$z = w\vec{x} + b$**

**$f_{(\vec{w},b)} (\vec{x}) = g(\vec{w} . \vec{x} + b) = g(z)$**

**$g(z) = 1/1+e^{-w\vec{x} + b}$**

the way to get probabilities from range of values (0 to 1) by sigmoid function
is by the threshold value, if the prediction is greater than or equal to 0.5 then it's probability of category 1 is more, similarly if the prediction is less than 0.5 then probability of category 0 is more.

$P(y=1|\vec{x};\vec{w},b)$

**Threshold = 0.5**

($f_{(\vec{w},b)} (\vec{x}) >= 0.5$)  =>  ($g(z) >= 0.5$) => ($g>=0$) => ($\vec{w} . \vec{x} + b >= 0$)

so, if $\vec{w} . \vec{x} + b >= 0$ then prediction() = 1.

similarly $\vec{w} . \vec{x} + b < 0$ then prediction() = 0.

**Decision Boundary**

at the Threshold value the Decison Boundary is defined at z=0

**$f_{(\vec{w},b)} (\vec{x}) = g(z) = g(\vec{w} . \vec{x} + b)$**

claculating the g(z) = 0, the decision boundary can be found, based on nthe type of polynomial used the decision Boundary varies from a line to many complex figures seperating defining the categories.

**using higher order polynomails for z**

by using the high order polynomials, the decision boundary can be more complicated not just a line $(\vec{w}.\vec{x} + b)$.

for complex data, higher order polynomial help better in defining a Decision Boundary.

without using the higher order polynomials you will always try to fit a line to every data and thats not a good model to predict with.


## Cost Function for Logistic Regression
cost function helps with the choice of parameters, that help in better fit the equation to the data.

for linear regressio we used squared error cost function,
this is not ideal for logistic regression, because the cost function graph will not be convex(non convex, bell/bowl shaped), unlike linear regression, this forms the graphs with multiple local minimas, and has higher chances of getting sucked in to one of the local minimas and not reaching global minima.

Defining a Loss Function for the Logstic Regression
$J(\vec{w},b) = 1/m ^{m}_{i=1} L(f_{\vec{w},b}(\vec{x}^{(i)}),y^{(i)})$

$L(f_{\vec{w},b}(\vec{x}^{(i)}),y^{(i)})$
This is the Loss function for one input of x and summation of m rows, from i=1 to m is the cost function(sum of all losses).

the Loss Function is defined as:

$L(f_{\vec{w},b}(\vec{x}^{(i)}),y^{(i)})$ = $-log(f_{\vec{w},b}(\vec{x}^{(i)}))$ if $y^{(i)} = 1$

$-log(1-f_{\vec{w},b}(\vec{x}^{(i)}))$  if $y^{(i)} = 0$


this Loss functions defined help to have the convex graph, that can reach the global minimum.

**simpler form of loss function**

$L(f_{\vec{w},b}(\vec{x}^{(i)}),y^{(i)})$ = $- (y^{(i)}) log(f_{\vec{w},b}(\vec{x}^{(i)})) - (1-y^{(i)})log(1-f_{\vec{w},b}(\vec{x}^{(i)}))$

because of logistic function(sigmoid) returns only 0 and 1, bt providing these y values the end equations are same as before.

## Cost Function for the Logistic Regression

$J(\vec{w},b) = 1/m ^{m}_{i=1} L(f_{\vec{w},b}(\vec{x}^{(i)}),y^{(i)})$

$L(f_{\vec{w},b}(\vec{x}^{(i)}),y^{(i)})$ = $- (y^{(i)}) log(f_{\vec{w},b}(\vec{x}^{(i)})) - (1-y^{(i)})log(1-f_{\vec{w},b}(\vec{x}^{(i)}))$

$J(\vec{w},b) = 1/m ^{m}_{i=1} (- (y^{(i)}) log(f_{\vec{w},b}(\vec{x}^{(i)})) - (1-y^{(i)})log(1-f_{\vec{w},b}(\vec{x}^{(i)})))$

$J(\vec{w},b) = -1/m ^{m}_{i=1} ( (y^{(i)}) log(f_{\vec{w},b}(\vec{x}^{(i)})) + (1-y^{(i)})log(1-f_{\vec{w},b}(\vec{x}^{(i)})))$

applied simplified Loss Function to the Cost Function of the Logistic Regression

##Gradient Descent
$w_j = w_j -  d/dw_j (J(\vec{w},b))$

$b = b -  d/db (J(\vec{w},b))$

applying the $J(\vec{w},b)$ cost function to it

the derivatives $d/dw_j(J(\vec{w},b))$ and $d/db(J(\vec{w},b))$ =>

$d/dw_j(J(\vec{w},b)) = 1/m ^{m}_{i=1} (f_{\vec{w},b}(\vec{x}^{(i)})-y^{(i)})(x^{(i)}_j)$

$d/db(J(\vec{w},b)) = 1/m ^{m}_{i=1} (f_{\vec{w},b}(\vec{x}^{(i)})-y^{(i)})$

these look same as of linear regression's Gradient descent. but the $f_{\vec{w},b}(\vec{x}^{(i)})$ value is the sigmoid function, not $(\vec{w}.\vec{x} + b).$

## Implementing Logistic Regression using scikit-learn
**implementing Logistic Regression for Titanic - Machine Learning from Disaster Competetion's data**



```python
import pandas as pd
train = pd.read_csv("/content/train (1).csv")
test = pd.read_csv("/content/test (1).csv")
display(train.head())
display(test.head())
```

```python
train.columns
```

```python
train.shape
```

```python
train.isna().sum()
```

```python
trainx = train.drop(['Cabin','Survived'], axis=1)
trainy = train['Survived']
testx = test.drop(['Cabin'], axis=1)
```

```python
trainx.info()
```

```python
cat = [col for col in trainx.columns if trainx[col].dtype == 'object']
num = [col for col in trainx.columns if trainx[col].dtype != 'object']
print([(i,trainx[i].nunique()) for i in cat])
```

```python
trainx = trainx.drop(['Name','Ticket','PassengerId'],axis=1)
cat = [col for col in trainx.columns if trainx[col].dtype == 'object']
num = [col for col in trainx.columns if trainx[col].dtype != 'object']
print(num)
print(cat)
```

```python
print([(i,trainx[i].nunique()) for i in cat])
print([i for i in num])
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler

cat_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
num_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', MinMaxScaler()) # Add MinMaxScaler here
])

preprocessor = ColumnTransformer(transformers=[
    ('cat', cat_transformer, cat),
    ('num', num_transformer, num)
])

model = Pipeline(steps=[
    ('preprocess',preprocessor),
    ('model',LogisticRegression())
])
```

```python
model.fit(trainx,trainy)
```

```python
testx.columns
```

```python
testx.drop(['PassengerId','Name','Ticket'],axis=1,inplace=True)
```

```python
predictions = model.predict(testx)
display(predictions)
```

```python
pd.DataFrame({'PassengerId':test['PassengerId'],'Survived':predictions}).to_csv('submission.csv',index=False)
```

```python
testy = pd.read_csv('/content/gender_submission (1).csv')
model.score(testx,testy['Survived'])
```

## Overfitting problem

1. when we try to implement a line(normal linear regression) to a curved data that doesnt actually fit correctly is a **underfitting model**.
its also reffered as "high-bias".
such models doesn't even work well with the training data it is learned from.

2. consider a model using quadratic polynomial to fit the data, this can fit the to the data quite well, and could generalize well to predict for the unknown data.
this is called **Generalization**, need this state of model in predicting near accurate values.

3. consider fitting a quadratic polynomial to the data , that might feel fitting very well to the training data, and could achieve almost 100% with the training data but such model will not perform well with unknown data, such state is called **Over-Fitting**.
it also reffered as "high-varience".

##Adressing Overfitting

1. use more data when high order polynimial is used (collect more data)
2. feature selection, selecting only a subset of features sacrificing the rest of the features, this help to not overfit to the data, with lot of featues and less data(entries, rows) the model will tend to overfit.
3. regularization
in point 2. the elemination of some features mean to making the parameters(w) to 0, instead of complete removal, the regularization makes the impact of those in the high order polynomial, with using the low values to them instead of high values, this reduces the impact of them in fitting to the data and helps not to overfit and generalize.


adding regularization to the cost function:

for linear regression, the cost function is:

$J(\vec{w},b) = 1/2m ^{m}_{i=1} (f_{\vec{w},b}(\vec{x}^{(i)}),y^{(i)})^2$

adding the regulaization part to it:

$J(\vec{w},b) = 1/2m ^{m}_{i=1} (f_{\vec{w},b}(\vec{x}^{(i)}),y^{(i)})^2 + /2m^{n}_{j=1}(w_j^2)$

**'' is called the Regularization parameter**

lamda is the hyperparameter, an appropriate value need to be chosen,a small value would not make any difference (overfits) , 0 value would completely make that part zero, and a large value will put more pressure of the regularization part causing problems, choosing a right value is important that generalizes the high order polynomial to the data.

we dont try to regularize b, because it doesnt make much difference.
