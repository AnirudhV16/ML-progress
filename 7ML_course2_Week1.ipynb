##working
a neural network has
1. input layer (input of features)
2. a hidden layer (layer of neurons)
each neuron of the hidden layer does mathematical computation, the hidden layer with many neurons have access to the input layer and can take any values.
**each neuron learns something from the input layer, and with a better set of features(outputs of hidden layer -> activations) to make final prediction**
3. the outputs of the hidden layer(activations) are inputs to other neuron like a logistic regression neuron to make the final probability prediction.

$\vec{x}(input vector) => HiddenLayer(Neurons) => \vec{a}(activation vector) => output Layer(some neurons)->prediction$

## simple neural network

consider a simple neural network of input layer, hidden layer with three neurons and output layer with one neuron(logistic regression).

layer(0) is input layer
layer(1) is hidden layer
layer(2) is output layer

1. the $\vec{x}$ is the input for the layer 1, for each neuron of the layer 1 all the values of the layer 0 are inputs

2. each neuron of layer 1, computing logistic regression,

    $g(z) -> g(\vec{w}_1.\vec{x} + b_1) = a_1$
    similarly for other two neurons

    $g(z) -> g(\vec{w}_2.\vec{x} + b_2) = a_2$

    $g(z) -> g(\vec{w}_3.\vec{x} + b_3) = a_3$

    all three are the outputs of the layer 1, the activations(sigmoid).

3. these a_1, a_2, a_3 are the activations(outputs) of the layer 1, which are input to the next layer 2

    layer 2 has one neuron(Logistic Regression),

    $g(z) -> g(\vec{w}_1.\vec{a}^{[1]} + b_1) = a$(output probability)

"a" is scalar single number, $a^{[2]}$

**$\vec{a}^{[1]}$ this activation is the output of the layer 1 and input to the layer 2**

**for all parameters and activation (output probability), super script like "[layer number]", is added**

## Forward Propogation
##Hand digit recognization (0,1) binary digit recognition

input: an image of (8*8) 64 blocks of brightness values, 255 for white and 0 for black.other numbers denote different shades of black and white.

a neural network of 3 layers:

 layer 0 -> input layer (vector of brightness values from the image)

 $layer1 -> hidden layer1, 25 neurons, input-> \vec{a}^{[0]}, output-> \vec{a}^{[1]}$

 $layer2 -> hidden layer2, 15 neurons, input-> \vec{a}^{[1]}, output-> \vec{a}^{[2]}$

 $layer3 -> output layer, 1 neuron, input-> \vec{a}^{[2]}, output-> a^{[3]}$

$a^{[3]}$ is the output scalar, optionally use threshold to find the output either 0 or 1.

the computation from left to right, propogating along all the activations, this is the forward propogation.
